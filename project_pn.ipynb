## AJUSTE DE LA CURVA MEDIANTE **REGRESIÓN LINEAL**

### Columna A.

```python
import pandas as pd
import numpy as np

# -----------------------------
# Datos
# -----------------------------
x = np.arange(0, 31)

y_A = [1.0,1.4,1.8,2.1,2.5,2.9,3.2,3.4,3.7,4.0,4.3,4.6,4.9,5.2,5.5,5.9,6.1,6.3,6.5,6.7,6.9,7.1,7.3,7.5,7.6,7.8,7.9,8.0,8.1,8.2,8.3]

# DataFrame
df = pd.DataFrame({
    'x': x,
    'y': y_A
})

# -----------------------------
# Inicialización
# -----------------------------
n = len(df)

sumx = 0
sumy = 0
sumxy = 0
sumx2 = 0
st = 0
sr = 0

# -----------------------------
# Sumatorias
# -----------------------------
for i in range(n):
    xi = df['x'][i]
    yi = df['y'][i]

    sumx = sumx + xi
    sumy = sumy + yi
    sumxy = sumxy + xi * yi
    sumx2 = sumx2 + xi * xi

# -----------------------------
# Promedios
# -----------------------------
xm = sumx / n
ym = sumy / n

# -----------------------------
# Coeficientes regresión
# -----------------------------
a1_num = (n * sumxy) - (sumx * sumy)
a1_den = (n * sumx2) - (sumx * sumx)

a1 = a1_num / a1_den
a0 = ym - a1 * xm

# -----------------------------
# Cálculo de errores
# -----------------------------
for i in range(n):
    xi = df['x'][i]
    yi = df['y'][i]

    st = st + (yi - ym)**2
    sr = sr + (yi - (a1 * xi + a0))**2

# -----------------------------
# Indicadores estadísticos
# -----------------------------
# Desviación estándar total
s = (st / (n - 1))**0.5

# Error estándar del estimado
syx = (sr / (n - 2))**0.5

# Coeficiente de determinación
r2 = (st - sr) / st

# Coeficiente de correlación
r = np.sqrt(r2)

# -----------------------------
# Resultados
# -----------------------------
print("Ecuación de regresión lineal:")
print(f"y = {a0:.4f} + {a1:.4f}x\n")

print("Indicadores estadísticos:")
print(f"Desviación estándar total (s): {s:.4f}")
print(f"Error estándar del estimado (Sy/x): {syx:.4f}")
print(f"Coeficiente de determinación (R^2): {r2:.4f}")
print(f"Coeficiente de correlación (r): {r:.4f}")
```

## AJUSTE DE LA CURVA MEDIANTE **REGRESIÓN POLINOMIAL DE SEGUNDO GRADO**

### Columna A.

```python
import pandas as pd
import numpy as np

# -----------------------------
# Datos
# -----------------------------
x = np.arange(0, 31)

y_A = [1.0,1.4,1.8,2.1,2.5,2.9,3.2,3.4,3.7,4.0,4.3,4.6,4.9,5.2,5.5,5.9,6.1,6.3,6.5,6.7,6.9,7.1,7.3,7.5,7.6,7.8,7.9,8.0,8.1,8.2,8.3]

df = pd.DataFrame({
    'x': x,
    'y': y_A
})

n = len(df)
order = 2  # grado del polinomio

# -----------------------------
# Construcción de la matriz
# -----------------------------
A = np.zeros((order+1, order+1))
B = np.zeros(order+1)

# -----------------------------
# Llenado de la matriz A
# -----------------------------
for i in range(order+1):
    for j in range(order+1):
        k = i + j
        suma = 0
        for l in range(n):
            xi = df['x'][l]
            suma = suma + xi**k
        A[i][j] = suma

# -----------------------------
# Llenado del vector B
# -----------------------------
for i in range(order+1):
    suma = 0
    for l in range(n):
        xi = df['x'][l]
        yi = df['y'][l]
        suma = suma + yi * (xi**i)
    B[i] = suma

# -----------------------------
# Resolución del sistema
# -----------------------------
coef = np.linalg.solve(A, B)

a0 = coef[0]
a1 = coef[1]
a2 = coef[2]

# -----------------------------
# Cálculo de errores
# -----------------------------
st = 0
sr = 0

ym = sum(df['y']) / n

for i in range(n):
    xi = df['x'][i]
    yi = df['y'][i]

    y_est = a0 + a1*xi + a2*(xi**2)

    st = st + (yi - ym)**2
    sr = sr + (yi - y_est)**2

# -----------------------------
# Indicadores estadísticos
# -----------------------------
s = (st / (n - 1))**0.5
syx = (sr / (n - (order + 1)))**0.5
r2 = (st - sr) / st
r = np.sqrt(r2)

# -----------------------------
# Resultados
# -----------------------------
print("Ecuación polinómica (grado 2):")
print(f"y = {a0:.6f} + {a1:.6f}x + {a2:.6f}x^2\n")

print("Indicadores estadísticos:")
print(f"Desviación estándar total (s): {s:.6f}")
print(f"Error estándar del estimado (Sy/x): {syx:.6f}")
print(f"Coeficiente de determinación (R^2): {r2:.6f}")
print(f"Coeficiente de correlación (r): {r:.6f}")
```

## AJUSTE DE LA CURVA MEDIANTE **REGRESIÓN EXPONENCIAL**.

### Columna A.

```python
import pandas as pd
import numpy as np

# -----------------------------
# Datos
# -----------------------------
x = np.arange(0, 31)

y_A = [1.0,1.4,1.8,2.1,2.5,2.9,3.2,3.4,3.7,4.0,4.3,4.6,4.9,5.2,5.5,5.9,6.1,6.3,6.5,6.7,6.9,7.1,7.3,7.5,7.6,7.8,7.9,8.0,8.1,8.2,8.3]

df = pd.DataFrame({
    'x': x,
    'y': y_A
})

n = len(df)

# -----------------------------
# Transformación logarítmica
# -----------------------------
df['Y'] = np.log(df['y'])

# -----------------------------
# Inicialización
# -----------------------------
sumx = 0
sumY = 0
sumxY = 0
sumx2 = 0

# -----------------------------
# Sumatorias
# -----------------------------
for i in range(n):
    xi = df['x'][i]
    Yi = df['Y'][i]

    sumx = sumx + xi
    sumY = sumY + Yi
    sumxY = sumxY + xi * Yi
    sumx2 = sumx2 + xi * xi

# -----------------------------
# Promedios
# -----------------------------
xm = sumx / n
Ym = sumY / n

# -----------------------------
# Coeficientes lineales
# -----------------------------
A1_num = (n * sumxY) - (sumx * sumY)
A1_den = (n * sumx2) - (sumx * sumx)

A1 = A1_num / A1_den
A0 = Ym - A1 * xm

# -----------------------------
# Volver al modelo original
# -----------------------------
a = np.exp(A0)
b = A1

# -----------------------------
# Cálculo de errores (en y real)
# -----------------------------
st = 0
sr = 0

ym = sum(df['y']) / n

for i in range(n):
    xi = df['x'][i]
    yi = df['y'][i]

    y_est = a * np.exp(b * xi)

    st = st + (yi - ym)**2
    sr = sr + (yi - y_est)**2

# -----------------------------
# Indicadores estadísticos
# -----------------------------
s = (st / (n - 1))**0.5
syx = (sr / (n - 2))**0.5
r2 = (st - sr) / st
r = np.sqrt(r2)

# -----------------------------
# Resultados
# -----------------------------
print("Modelo exponencial:")
print(f"y = {a:.6f} * e^({b:.6f}x)\n")

print("Indicadores estadísticos:")
print(f"Desviación estándar total (s): {s:.6f}")
print(f"Error estándar del estimado (Sy/x): {syx:.6f}")
print(f"Coeficiente de determinación (R^2): {r2:.6f}")
print(f"Coeficiente de correlación (r): {r:.6f}")
```

```python
import pandas as pd
import numpy as np

# -----------------------------
# Datos
# -----------------------------
x = np.arange(0, 31)
y_A = np.array([1.0,1.4,1.8,2.1,2.5,2.9,3.2,3.4,3.7,4.0,4.3,4.6,4.9,5.2,5.5,5.9,6.1,6.3,6.5,6.7,6.9,7.1,7.3,7.5,7.6,7.8,7.9,8.0,8.1,8.2,8.3])
y_B = np.array([1.0,1.3,1.6,1.9,2.1,2.4,2.6,2.9,3.1,3.3,3.6,3.8,4.1,4.3,4.6,4.8,5.0,5.2,5.4,5.6,5.8,6.0,6.2,6.4,6.5,6.7,6.8,7.0,7.1,7.2,7.3])
y_C = np.array([1.0,1.2,1.3,1.5,1.6,1.7,1.8,2.0,2.1,2.2,2.3,2.5,2.6,2.7,2.8,3.0,3.1,3.2,3.3,3.4,3.5,3.6,3.7,3.8,3.9,4.0,4.1,4.2,4.3,4.4,4.5])

# Se crea una lista para almacenar los datos estadísticos.
statistics_data = {}

# ----------------------------- #
# --- Regresión Lineal ---      #
# ----------------------------- #

# Se define función para calcular los estadísticos de la regresión lineal.
def calculate_linear_regression_stats(x_data, y_data):
    n = len(x_data)

    if n < 2:
        return np.nan, np.nan, np.nan, np.nan

    sumx = np.sum(x_data)
    sumy = np.sum(y_data)
    sumxy = np.sum(x_data * y_data)
    sumx2 = np.sum(x_data * x_data)
    promy = sumy / n
    promx = sumx / n

    # Coeficientes regresión
    a1_num = (n * sumxy) - (sumx * sumy)
    a1_den = (n * sumx2) - (sumx * sumx)

    if a1_den == 0:
        return np.nan, np.nan, np.nan, np.nan

    # Pendiente e intercepto
    a1 = a1_num / a1_den
    a0 = (sumy / n) - a1 * (sumx / n)

    # Cálculo de errores
    st = np.sum((y_data - np.mean(y_data))**2)
    y_pred = a1 * x_data + a0
    sr = np.sum((y_data - y_pred)**2)

    s = (st / (n - 1))**0.5 if (n-1) > 0 else np.nan
    syx = (sr / (n - 2))**0.5 if (n-2) > 0 else np.nan
    r2 = (st - sr) / st if st > 0 else np.nan
    r = np.sqrt(r2) if not np.isnan(r2) and r2 >= 0 else np.nan

    return s, syx, r2, r

# Calcular para cada columna.
s_A_lin, syx_A_lin, r2_A_lin, r_A_lin = calculate_linear_regression_stats(x, y_A)
statistics_data[('A', 'Lineal')] = {'Desviación estándar total (s)': s_A_lin, 'Error estándar del estimado (Sy/x)': syx_A_lin, 'Coeficiente de determinación (R^2)': r2_A_lin, 'Coeficiente de correlación (r)': r_A_lin}

s_B_lin, syx_B_lin, r2_B_lin, r_B_lin = calculate_linear_regression_stats(x, y_B)
statistics_data[('B', 'Lineal')] = {'Desviación estándar total (s)': s_B_lin, 'Error estándar del estimado (Sy/x)': syx_B_lin, 'Coeficiente de determinación (R^2)': r2_B_lin, 'Coeficiente de correlación (r)': r_B_lin}

s_C_lin, syx_C_lin, r2_C_lin, r_C_lin = calculate_linear_regression_stats(x, y_C)
statistics_data[('C', 'Lineal')] = {'Desviación estándar total (s)': s_C_lin, 'Error estándar del estimado (Sy/x)': syx_C_lin, 'Coeficiente de determinación (R^2)': r2_C_lin, 'Coeficiente de correlación (r)': r_C_lin}

# ----------------------------- #
# --- Regresión Polinomial (Grado 2) --- #
# ----------------------------- #

# Se define función para calcular los estadísticos de la regresión polinomial.
def calculate_poly_regression_stats(x_data, y_data, order):
    n = len(x_data)
    if n < order + 1:
        return np.nan, np.nan, np.nan, np.nan

    # Ajuste polinómico
    coef = np.polyfit(x_data, y_data, order)

    # Cálculo de errores
    st = np.sum((y_data - np.mean(y_data))**2)
    y_pred = np.polyval(coef, x_data)
    sr = np.sum((y_data - y_pred)**2)

    s = (st / (n - 1))**0.5 if (n-1) > 0 else np.nan
    syx = (sr / (n - (order + 1)))**0.5 if (n - (order + 1)) > 0 else np.nan
    r2 = (st - sr) / st if st > 0 else np.nan
    r = np.sqrt(r2) if not np.isnan(r2) and r2 >= 0 else np.nan

    return s, syx, r2, r

# Calcular para cada columna.
s_A_quad, syx_A_quad, r2_A_quad, r_A_quad = calculate_poly_regression_stats(x, y_A, 2)
statistics_data[('A', 'Cuadrática')] = {'Desviación estándar total (s)': s_A_quad, 'Error estándar del estimado (Sy/x)': syx_A_quad, 'Coeficiente de determinación (R^2)': r2_A_quad, 'Coeficiente de correlación (r)': r_A_quad}

s_B_quad, syx_B_quad, r2_B_quad, r_B_quad = calculate_poly_regression_stats(x, y_B, 2)
statistics_data[('B', 'Cuadrática')] = {'Desviación estándar total (s)': s_B_quad, 'Error estándar del estimado (Sy/x)': syx_B_quad, 'Coeficiente de determinación (R^2)': r2_B_quad, 'Coeficiente de correlación (r)': r_B_quad}

s_C_quad, syx_C_quad, r2_C_quad, r_C_quad = calculate_poly_regression_stats(x, y_C, 2)
statistics_data[('C', 'Cuadrática')] = {'Desviación estándar total (s)': s_C_quad, 'Error estándar del estimado (Sy/x)': syx_C_quad, 'Coeficiente de determinación (R^2)': r2_C_quad, 'Coeficiente de correlación (r)': r_C_quad}

# ----------------------------- #
# --- Regresión Exponencial --- #
# ----------------------------- #

def calculate_exp_regression_stats(x_data, y_data):
    n = len(x_data)
    if n < 2:
        return np.nan, np.nan, np.nan, np.nan

    # Asegurarse de que no haya valores <= 0 para logaritmo
    if np.any(y_data <= 0):
        return np.nan, np.nan, np.nan, np.nan # Or handle appropriately

    log_y = np.log(y_data)

    # Regresión lineal sobre datos transformados
    coef_exp = np.polyfit(x_data, log_y, 1)
    A1 = coef_exp[0]
    A0 = coef_exp[1]

    a = np.exp(A0)
    b = A1

    # Cálculo de errores (en y real)
    ym = np.mean(y_data)
    st = np.sum((y_data - ym)**2)
    y_pred = a * np.exp(b * x_data)
    sr = np.sum((y_data - y_pred)**2)

    s = (st / (n - 1))**0.5 if (n-1) > 0 else np.nan
    syx = (sr / (n - 2))**0.5 if (n-2) > 0 else np.nan
    r2 = (st - sr) / st if st > 0 else np.nan
    r = np.sqrt(r2) if not np.isnan(r2) and r2 >= 0 else np.nan

    return s, syx, r2, r

# Calcular para cada columna
s_A_exp, syx_A_exp, r2_A_exp, r_A_exp = calculate_exp_regression_stats(x, y_A)
statistics_data[('A', 'Exponencial')] = {'Desviación estándar total (s)': s_A_exp, 'Error estándar del estimado (Sy/x)': syx_A_exp, 'Coeficiente de determinación (R^2)': r2_A_exp, 'Coeficiente de correlación (r)': r_A_exp}

s_B_exp, syx_B_exp, r2_B_exp, r_B_exp = calculate_exp_regression_stats(x, y_B)
statistics_data[('B', 'Exponencial')] = {'Desviación estándar total (s)': s_B_exp, 'Error estándar del estimado (Sy/x)': syx_B_exp, 'Coeficiente de determinación (R^2)': r2_B_exp, 'Coeficiente de correlación (r)': r_B_exp}

s_C_exp, syx_C_exp, r2_C_exp, r_C_exp = calculate_exp_regression_stats(x, y_C)
statistics_data[('C', 'Exponencial')] = {'Desviación estándar total (s)': s_C_exp, 'Error estándar del estimado (Sy/x)': syx_C_exp, 'Coeficiente de determinación (R^2)': r2_C_exp, 'Coeficiente de correlación (r)': r_C_exp}

# Crear DataFrame con índice multinivel
df_stats = pd.DataFrame.from_dict(statistics_data, orient='index')
df_stats.index.names = ['Columna', 'Modelo']

print("Tabla de Estadísticos de Regresión por Columna y Modelo:")
display(df_stats)
```
